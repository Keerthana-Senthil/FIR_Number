```python
import pandas as pd
```

```python
df=pd.read_csv('./DDP_fir_2022.csv')
```

```python
df.dropna(inplace = True)
```

```python
dfnew=df['fir_number'].str.split('/')
df["correctness"]= ""
print(dfnew)

```

```python
for i in range(len(dfnew)):
	p=1
	if len(dfnew[i])== 2:
		for j in range(len(dfnew[i])):
			if len(dfnew[i][j])<=4:
				u = int(dfnew[i][j].isnumeric())
				p = p*u
			else:
				p = 0
	else:
		p=0
	df["correctness"][i] = p
	print('Entry number',i,'Value',p)
```

```python
print(df)
```

```python
import csv
with open('./output.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    header = ["fir_number","correctness"]
    writer.writerow(header)
    for i in range(len(df)):
        List = [df["fir_number"][i],df["correctness"][i]]
        writer.writerow(List)
```
